```python
from PIL import Image
from PIL import ImageEnhance
#原始图像
image = Image.open("data\\test.png")
image.show()
#亮度增强
enh_bri = ImageEnhance.Brightness(image)
brightness = 1.5
image_brightened = enh_bri.enhance(brightness)
image_brightened.show()
```

```python
#对比度增强
from PIL import Image
from PIL import ImageEnhance
from PIL import ImageFilter
#原始图像
image = Image.open("data\\test.png")
image.show()
#对比度增强
enh_con = ImageEnhance.Contrast(image)
contrast = 5.0
image_contrasted = enh_con.enhance(contrast)
image_contrasted.show()
#锐度增强
enh_sha = ImageEnhance.Sharpness(image)
sharpness = 5.0
image_sharped = enh_sha.enhance(sharpness)
image_sharped.show()
#图像模糊
im = Image.open("data\\test.png")
im_blur = im.filter(ImageFilter.BLUR)
im_blur.show()
#轮廓提取
im = Image.open("data\\test.png")
im_contour = im.filter(ImageFilter.CONTOUR)
im_contour.show()
```

```python
#图像分割
# 图片二值化
from PIL import Image
img = Image.open('data\\test.png')
# 模式 L”为灰色图像，它的每个像素用 8 个 bit 表示，0 表示黑，255 表示白，其他数字表示不同的灰度。
Img = img.convert('L')
Img.save("data\\test1.jpg")
# 自定义灰度界限threhold，大于这个值为黑色，小于这个值为白色
threshold = 100
table = []
# f=lambda x:table.append(0) if i<threshold else table.append(1)
for i in range(256):
    if i < threshold:
        table.append(0)
    else:
        table.append(1)
# 图片二值化
photo = Img.point(table, '1')
photo.show()
```
